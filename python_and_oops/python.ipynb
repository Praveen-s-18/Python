- **Conditional statements**

```python
## if, elif, else, nested if, match case

a = 18
b = 20
c = 15

if(a > 0):
    print("+ve")
elif (a < 0):
    print("-ve")
else:
    print("0")
    
    
if a > b and a > c:
    print("a is the largest")
elif b > a and b > c:
    print("b is the largest")
else:
    print("c is the largest")
    
    
lst = []
if lst:
    print("not empty")
else:
    print("empty")
    
    
# Nested if

if a > 0:
    if a % 2 == 0:
        print("+ve even")
    else:
        print("+ve odd")

elif a < 0:
    if a % 2 == 0:
        print("-ve even")
    else:
        print("-ve odd")

else:
    print("0")
    

# match case

x = 4
match x:
    case 1:
        print("one")
    case 2:
        print("two")
    case 3:
        print("three")
    case _:
        print("invalid")
```

- **looping statements**

```python
## for loop   -> iterate over iterable / sequence (list, tuple, set, string, dictionary,range object)


lst = [1, 2, 3, 4, 5]
for i in lst:
    print(i, end = " ")      
print()                                             # for loop used when number of iterations is known

for i in range(5):
    print(i, end = " ")
print()
    
for i in range(1, 10, 2):  # start, stop, step
    print(i, end = " ")
print()

# nested for loop
for i in range(1, 4):
    for j in range(1, 4):
        print(i, j)

# for else
for i in range(5):
    if i == 6:
        print("found 3")
        break
else:
    print("loop completed without break")
```

```python
 ## While loop -> execute block of code repeat until condition is False

i = 0
while i < 5:
    print(i, end = " ")
    i += 1
    

# nested while loop

i = 1
while i < 3:                            ## while loop used when number of iterations is not known
    j = 1
    while j < 3:
        print(i, j)
        j += 1
    i += 1

## while else

i = 0
while i < 5:
    if i == 6:
        print("found 3")
        break
    i += 1
else:
    print("loop completed without break")
    
```

- **jump statements**  -->  break, continue, pass

```python
# Break

for i in range(5):
    if i == 3:
        print("found 3")  # terminate loop
        break

# continue

for i in range(10):
    if i % 2 == 0:          # skip current iteration
        continue
    print(i, end = " ")


# pass
for i in range(5):      # do nothing , placeholder for future codes
    pass
print()


# return

def add(a, b):
    return a + b


# yield

def test():
    yield 1                 # pass 1
    print("Hello")          # returns value ,pauses function, saves state, resumes later
    yield 2                 # pass 2        

t = test()

print(next(t))
print(next(t))
yield 

```

```python
## Lambda with built in functions

# map()

l = [10, 20, 30, 40, 50]
res = map(lambda x: x // 2, l)
print(res)
print(list(res))


# filter()

l = [10, 15, 20, 25, 30]
res = filter(lambda x: x % 2 == 0, l)  #  --> lambda x: True if x % 2 == 0 else False
print(res)                             # only true values included
print(list(res))


# sorted()
l = [5, 2, 9, 1, 5, 6]
res = sorted(l, key = lambda x: x)
print(res)


res = sorted(l, key = lambda x: x, reverse = True)  # sort in descending order
print(res)
```

- `class method`

    - *Access and modify class variable*
    - use `@classmethod` decorator

```python
class Student:
    
    college = "ABC"
    
    @classmethod
    def show_college(cls):
        print(cls.college)
        
    @classmethod
    def update_college(cls, new_name):
        cls.college = new_name             
    
    

Student.show_college()
Student.update_college("XYZ")           # without creating object access class methods
Student.show_college()
```

#### ***Encapsulation***

```python
class Employee:
    
    def __init__(self, salary):
        self.__salary = salary

    def get_salary(self):
        return self.__salary
    
    def set_salary(self, new_salary):
        self.__salary = new_salary

e = Employee(50000)
print(e.get_salary())
e.set_salary(60000)
print(e.get_salary())
```

#### ***Inheritance***

- `single inheritance`

```python
class Person:
    
    def __init__(self, name):
        self.name = name

    def show_name(self):
        print("Name:", self.name)


class Student(Person):

    def __init__(self, name, roll):
        super().__init__(name)
        self.roll = roll

    def show_student(self):
        print("Roll:", self.roll)
        print("access parent class variable : ", self.name) 
        
        
s = Student("Alice", 22)
s.show_student()
s.show_name()
```

- `multilevel inheritance`

```python
class Person:
    def __init__(self, name):
        print("Person constructor")
        self.name = name
        
    def show_name(self):
        print("Name:", self.name)


class Student(Person):
    def __init__(self, name, roll):
        print("Student constructor")
        super().__init__(name)
        self.roll = roll
        
    def show_student(self):
        print("Roll:", self.roll)
        print("access parent class variable : ", self.name)


class Result(Student):
    def __init__(self, name, roll, marks):
        print("Result constructor")
        super().__init__(name, roll)
        self.marks = marks

    def show(self):
        print(self.name, self.roll, self.marks)
        
        
r = Result("Alice", 22, 85)
r.show()
```

- `Hierarchical inheritance`

```python
class Person:
    
    def __init__(self, name):
        self.name = name
        
    def show_name(self):
        print("Name:", self.name)
        
class Student(Person):
    
    def __init__(self, name, rollno):
        super().__init__(name)
        self.rollno = rollno
        
    def show_student(self):
        print(f"Roll No:{self.rollno}, Name: {self.name}")
    
    
class Teacher(Person):
    
    def __init__(self, name, subject):
        super().__init__(name)
        self.subject = subject
        
    def show_teacher(self):
        print(f"teacher : {self.name}, subject : {self.subject}")
        

s = Student("Alice", 101)
t = Teacher("Bob", "Math")

s.show_name()
s.show_student()
print()

t.show_name()
t.show_teacher()
```

`multiple inheritance`

```python
class Person:
    
    def __init__(self, name):
        self.name = name

    def show_name(self):
        print("Name:", self.name)


class Academic:

    def __init__(self, marks):
        self.marks = marks

    def show_marks(self):
        print("Marks:", self.marks)


class Student(Person, Academic):    

    def __init__(self, name, marks, roll):
        Person.__init__(self, name)
        Academic.__init__(self, marks)
        self.roll = roll

    def show_student(self):
        print("Roll:", self.roll)
        
        
s = Student("Alice", 85, 1)
s.show_name()      
s.show_marks()     
s.show_student()  
```

- `hybrid inheritance`

```python
class Appliance:
    
    def __init__(self, brand):
        self.brand = brand
        
    def turn_on(self):
        print(f"{self.brand} turned on")


class Washing_machine(Appliance):
    
    def wash(self):
        print(f"{self.brand} washes clothes")


class Refrigerator(Appliance):
    
    def cool(self):
        print(f"{self.brand} cools food")


class Smart_appliance(Washing_machine, Refrigerator):
    
    def smart_feature(self):
        print(f"{self.brand} has smart feature")



s = Smart_appliance("LG")

s.turn_on()         
s.wash()            
s.cool()            
s.smart_feature()  
```

### ***Polymorphism***

```python
# python not method overloading

class Calculator:
    
    def add(self, a, b, c=0):       # default argument
        print(a + b + c)


c = Calculator()

c.add(10, 20)
c.add(10, 20, 30)
```

```python
class Calculator:
    
    def add(self, *args):           # variable length arguments
        i = 0
        for num in args:
            i += num
        print(i)
        
c = Calculator()
c.add(10, 20)
c.add(10, 20, 30)
c.add(10, 20, 30, 40)
```

```python
class Animal:
    def sound(self):
        print("Some generic sound")

class Dog(Animal):
    def sound(self):
        print("Bark")

class Cat(Animal):
    def sound(self):
        print("Meow")


animals = [Dog(), Cat(), Animal()]
for animal in animals:
    animal.sound()
```

### ***Abstraction***

```python
from abc import ABC, abstractmethod

class Payment(ABC):
    
    @abstractmethod
    def pay(self, amount):
        pass
    
    def payment_info(self):
        print("This is a payment service") 

class CreditCardPayment(Payment):
    
    def pay(self, amount):
        print(f"Paid ₹{amount} using Credit Card")


class UPIPayment(Payment):
    
    def pay(self, amount):
        print(f"Paid ₹{amount} using UPI")


p1 = CreditCardPayment()
p2 = UPIPayment()
p1.payment_info()

p1.pay(500)
p2.pay(1000)
```
