```python
import os
from autogen import ConversableAgent

config_list = [
    {
        "model": "llama-3.3-70b-versatile",
        "api_key": os.environ.get("GROQ_API_KEY"),
        "base_url": "https://api.groq.com/openai/v1",
        "api_type": "openai",
        "price": [0, 0],
        "temperature": 0.7,
    }
]

agent = ConversableAgent(
    "chatbot",
    llm_config={"config_list": config_list},
    code_execution_config = False,
    function_map = None,
    human_input_mode = "NEVER",
)
```

```python
reply = agent.generate_reply(messages=[{"content": "Explain RAG pipeline step by step.", "role": "user"}])
print(reply)
```

```python

cathy = ConversableAgent(
    "cathy",
    system_message="Your name is Cathy and you are a part of a duo of comedians.",
    llm_config={"config_list":config_list},
    human_input_mode="NEVER",  # Never ask for human input.
)

joe = ConversableAgent(
    "joe",
    system_message="Your name is Joe and you are a part of a duo of comedians.",
    llm_config={"config_list": config_list},
    human_input_mode="NEVER",  # Never ask for human input.
)
```

```python
result = joe.initiate_chat(cathy, message="Cathy, tell me a joke.", max_turns=2)
```

```python
# terminate conversation

teacher = ConversableAgent(
    "teacher",
    system_message="You ask simple math questions and correct answers.",
    llm_config={"config_list": config_list},
    human_input_mode="NEVER",
     max_consecutive_auto_reply=2,
)

student = ConversableAgent(
    "student",
    system_message="You try to answer math questions.",
    llm_config={"config_list": config_list},
    human_input_mode="NEVER",
    max_consecutive_auto_reply=2,
)

student.initiate_chat(teacher, message="Ask me a question.", max_turns=4)
```

```python
teacher = ConversableAgent(
    "teacher",
    system_message=(
        "You ask simple math questions one by one. "
        "After 1 question, end the conversation by saying 'good bye'."
    ),
    llm_config={"config_list": config_list},
    human_input_mode="NEVER",
    max_consecutive_auto_reply=3,  
    is_termination_msg=lambda msg: "good bye" in msg["content"].lower(),  # 🔹 Smart stop
)

student = ConversableAgent(
    "student",
    system_message="You answer math questions.",
    llm_config={"config_list": config_list},
    human_input_mode="NEVER",
    max_consecutive_auto_reply=3,  
    is_termination_msg=lambda msg: "good bye" in msg["content"].lower(),  # 🔹 Smart stop
)

student.initiate_chat(teacher, message="Ask me a math question.")
```

```python
agent_with_number = ConversableAgent(
    "agent_with_number",
    system_message="You are the host of a guess-my-number game. You have the number 15. "
                   "Reply 'too high' or 'too low'. If the guess is 15, say 'Correct! 15'.",
    llm_config={"config_list": config_list},
    is_termination_msg=lambda msg: "15" in msg["content"], 
    human_input_mode="ALWAYS",  
)

agent_guess_number = ConversableAgent(
    "agent_guess_number",
    system_message="You are a player trying to guess a number between 1 and 25. "
                   "Use the feedback 'too high' or 'too low' to find it.",
    llm_config={"config_list": config_list},
    human_input_mode="NEVER", 
)

# 4. Start the Chat
result = agent_with_number.initiate_chat(
    agent_guess_number,
    message="I have a number between 1 and 25. Guess it!",
)
```
