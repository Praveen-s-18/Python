```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
        
class LinkedList:
    def __init__(self):
        self.head = None
        
    def insert_at_end(self, data):
        new_node = Node(data)
        
        if self.head is None:
            self.head = new_node
        else:
            temp = self.head
            while temp.next:
                temp = temp.next
            temp.next = new_node
            
    def insert_at_beg(self, data):
        new_node = Node(data)
        if self.head is None:
            self.head = new_node
        
        else:
            new_node.next = self.head
            self.head = new_node
    
    def insert_at_pos(self, pos, data):
        new_node = Node(data)
        if pos == 1:
            new_node.next = self.head
            self.head = new_node
            
        else:
            temp = self.head
            for i in range(1, pos -1):
                temp = temp.next
            new_node.next = temp.next
            temp.next = new_node
            
    def middle_element(self):
        slow = self.head
        fast = self.head
        
        while fast and fast.next:
            slow = slow.next
            fast = fast.next.next
        return slow.data
    
    def ll_reverse(self):
        prev = None
        curr = self.head
        next = None
        while curr:
            next = curr.next
            curr.next = prev
            prev = curr
            curr = next
        self.head = prev
                
        
    def display(self):
        temp = self.head
        while temp:
            print(temp.data, end=" -> ")
            temp = temp.next
        print("None")
        
ll = LinkedList()
n = int(input("No.of nodes"))

for i in range(n):
    data = int(input("value : "))
    ll.insert_at_end(data)

ll.display()


pos = int(input("Enter the pos : "))
val = int(input("Enter the value to be inserted"))
ll.insert_at_pos(pos, val)

ll.display()


l2 = LinkedList()
n = int(input("No.of nodes"))
for i in range(n):
    data = int(input("value : "))
    l2.insert_at_beg(data)
l2.display()
ll.ll_reverse()
ll.display()
ll.middle_element()
```

### ***Stack***

- `using array`

```python
class Stack:
    
    def __init__(self, size):
        self.stack = [None]*size
        self.size = size
        self.top = -1
        
    
    def push(self):
        if self.top == self.size - 1:
            print("Stack overflow")
        else:
            self.top += 1
            n = int(input("Eneter element : "))
            self.stack[self.top] = n
            print("Inserted")
    
    def pop(self):
        if self.top == -1:
            print("Stack underflow")
        else:
            self.stack[self.top] = None
            self.top -= 1
    
    def show(self):
        if self.top == -1: print("Underflow")
        else:
            for i in range(self.top,-1,-1):
                print(self.stack[i], end = " ")
            print()
    
    def peek(self):
        if self.top == -1:
            print("underflow")
        else:
            print(self.stack[self.top])
            


stack = Stack(5)

while True:
    choice = int(input())
    if choice == 1:
        stack.push()
    elif choice == 2:
        stack.pop()
    elif choice == 3:
        stack.show()
    elif choice == 4:
        stack.peek()
    else:
        break

        
```

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class Stack:
    def __init__(self):
        self.top = None

    def push(self, data):
        new_node = Node(data)
        new_node.next = self.top
        self.top = new_node
        print(f"{data} pushed into stack")

    def pop(self):
        if self.top is None:
            print("Underflow! Stack is empty.")
        else:
            pop_element = self.top.data
            self.top = self.top.next
            print("Popped:", pop_element)

    def peek(self):
        if self.top is None:
            print("Underflow! Stack is empty.")
        else:
            print("Top element is:", self.top.data)

    def show(self):
        if self.top is None:
            print("Underflow! Stack is empty.")
        else:
            print("Stack elements:")
            temp = self.top
            while temp:
                print(temp.data, end=" -> ")
                temp = temp.next
            print("None")

stack = Stack()

while True:

    choice = int(input("\nEnter your choice: "))

    if choice == 1:
        value = int(input("Enter value to push: "))
        stack.push(value)
    elif choice == 2:
        stack.pop()
    elif choice == 3:
        stack.show()
    elif choice == 4:
        stack.peek()
        
    else:
        break
```

### ***Queue***

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
        
class Queue:
    def __init__(self):
        self.front = None
        self.rear = None
    
    def enqueue(self, data):
        new_node = Node(data)
        if self.rear is None:
            self.front = self.rear = new_node
        else:
            self.rear.next = new_node
            self.rear = new_node
    
    
    def dequeue(self):
        if self.front is None:
            print("Empty")
        else:
            dequeued_element = self.front.data
            self.front = self.front.next
            print("Element dequeued : ",dequeued_element)
            if self.front is None:
                self.rear = None
    
    
    def peek(self):
        if self.front is None:
            print("empty")
        else:
            print("peek : ",self.front.data)
    
    def display(self):
        if self.front is None:
            print("Empty")
        temp = self.front
        while temp:
            print(temp.data, end = " ")
            temp = temp.next
    
queue = Queue()

while True:
        
    choice = int(input("\nEnter your choice: "))
    if choice == 1:
        value = int(input("Enter value to enqueue: "))
        queue.enqueue(value)
    elif choice == 2:
        queue.dequeue()
    elif choice == 3:
        queue.display()
    elif choice == 4:
        queue.peek()
    else:
        break

            
        
        
```

### ***Hash Map***

```python
class HashMap:
    
    def __init__(self, size=10):
        self.size = size
        self.table = [[] for _ in range(size)]

    def _hash(self, key):
        return hash(key) % self.size

    def put(self, key, value):
        index = self._hash(key)
        bucket = self.table[index]

        for i, (k, v) in enumerate(bucket):
            if k == key:
                bucket[i] = (key, value)
                return
        bucket.append((key, value))

    def get(self, key):
        index = self._hash(key)

        for k, v in self.table[index]:
            if k == key:
                return v
        return None

    def delete(self, key):
        index = self._hash(key)

        for i, (k, v) in enumerate(self.table[index]):
            if k == key:
                self.table[index].pop(i)
                return

    def display(self):
        print(self.table)



hm = HashMap()

hm.put("name", "Praveen")
hm.put("age", 22)

print(hm.get("name"))

hm.delete("age")

hm.display()
```

### ***Sorting algorithms***

- `Bubble Sort`

```python
arr = [5, 3, 8, 4, 2]
n = len(arr)

for i in range(n):
    swap = False
    for j in range(n - i - 1):
        if arr[j] < arr[j + 1]:
            arr[j], arr[j+1] = arr[j+1], arr[j]
            swap = True
            
    if not swap:
        break
    
print(arr)
```

- `Selection Sort`

```python
arr = [5, 3, 8, 4, 2]
n = len(arr)

for i in range(n):
    min_ind = i
    for j in range(i + 1, n):
        if arr[j] < arr[min_ind]:
            min_ind = j
    arr[i], arr[min_ind] = arr[min_ind], arr[i]
print(arr)
            
```

- `Insertion sort`

```python
arr = [5, 3, 8, 4, 2]
n = len(arr)

for i in range(1, n):
    key = arr[i]
    j = i - 1
    while j >= 0 and arr[j] > key:
        arr[j + 1] = arr[j]
        j -= 1
    arr[j + 1] = key
print(arr)
        
```

- `Merge sort`

```python
def merge(arr, low, mid, high):
    temp = []
    left = low
    right = mid + 1

    while left <= mid and right <= high:
        if arr[left] <= arr[right]:
            temp.append(arr[left])
            left += 1
        else:
            temp.append(arr[right])
            right += 1

    while left <= mid:
        temp.append(arr[left])
        left += 1

    while right <= high:
        temp.append(arr[right])
        right += 1

    for i in range(low, high + 1):
        arr[i] = temp[i - low]


def merge_sort(arr, low, high):
    if low >= high:
        return

    mid = (low + high) // 2

    merge_sort(arr, low, mid)
    merge_sort(arr, mid + 1, high)
    merge(arr, low, mid, high)


arr = [5, 2, 8, 4, 1]

merge_sort(arr, 0, len(arr) - 1)

print(arr)
```

- `Binary search`

```python
def binary_search(arr):
    
    low = 0
    high = len(arr) - 1
    while low <= high:
        mid = (low + high) // 2
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            low = mid + 1
        else:
            high = mid - 1
    return -1

arr = [1, 2, 3, 4, 5]
target = 3
result = binary_search(arr)
if result != -1:
    print("Element found at index:", result)
else:
    print("Element not found in the array.")
    
```

### ***Recursion***

- `factorial`

```python
def factorial(n):
    
    if n == 0 or n == 1:
        return 1
    
    return n * factorial(n - 1)

print(factorial(5))
```

- `fibonacci`

```python
def fibonacci(n):
    
    if(n <= 1):
        return n
    
    return fibonacci(n - 1) + fibonacci(n - 2)

print(fibonacci(5))
```

- `Reverse array`

```python
def reverse(arr, left, right):
    
    if left >= right:
        return
    
    arr[left], arr[right] = arr[right], arr[left]
    return reverse(arr, left + 1, right - 1)

arr = [1, 2, 3, 4, 5, 6]
reverse(arr, 0, len(arr) - 1)
print(arr)
```

```python
def fib(n):
    
    dp = [0] * (n + 1)
    dp[0] = 0
    dp[1] = 1
    for i in range(2, n + 1):
        dp[i] = dp[i -1] +  dp[i - 2]
        
    return dp[n]

n = 6
print(fib(n))
```

```python
from collections import deque
class Node:
    
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None
        
class BinaryTree:
    
    def insert(self, root, data):
        
        if root is None:
            return Node(data)
        if data < root.data:
            root.left = self.insert(root.left, data) 
        elif data > root.data:
            root.right = self.insert(root.right, data)
            
        return root
    
    def preorder(self, root):
        
        if root is None:
            return
        
        print(root.data, end = " ")
        self.preorder(root.left)
        self.preorder(root.right)
    
    
    def inorder(self, root):
        
        if root is None:
            return
        
        self.inorder(root.left)
        print(root.data, end = " ")
        self.inorder(root.right)
        
    
    def postorder(self, root):
        
        if root is None:
            return
        
        self.postorder(root.left)
        self.postorder(root.right)
        print(root.data, end = " ")
        
    def levelorder(self, root):
        if root is None:
            return
        
        queue = deque()
        queue.append(root)
        res = []
        
        while queue:
            temp = []
            n = len(queue)
            
            for i in range(n):
                
                node = queue.popleft()
                temp.append(node.data)
                if node.left:
                    queue.append(node.left)
                if node.right:
                    queue.append(node.right)
            res.append(temp)
        return res
```

```python
bt = BinaryTree()
root = None

el = list(map(int, input().split()))
for i in el:
    root = bt.insert(root, i)
    
    
print("preorder : ", bt.preorder(root))
print("inorder : ", bt.inorder(root))
print("postorder : ", bt.postorder(root))
print("level order : ", bt.levelorder(root))
```

### ***Graph***

```python
class Graph:
    
    def __init__(self):
        self.graph = {}
        
    def add_vertex(self, vertex):
        
        if vertex not in self.graph:
            self.graph[vertex] = []
            
    
    def add_edge(self, vertex1, vertex2):
        
        if vertex1 not in self.graph:
            self.add_vertex(vertex1)
        if vertex2 not in self.graph:
            self.add_vertex(vertex2)
        
        self.graph[vertex1].append(vertex2)
        self.graph[vertex2].append(vertex1)
        
    
    def display(self):
        
        for node in self.graph:
            print(node, "->", self.graph[node])


g = Graph()

g.add_vertex(0)
g.add_vertex(1)
g.add_vertex(2)
g.add_vertex(3)

g.add_edge(0,1)
g.add_edge(0,2)
g.add_edge(1,3)
g.add_edge(2,3)

g.display()
```

```python
class Graph:
    
    def __init__(self, vertices):
        self.vertices = vertices
        self.matrix = [[0]*vertices for _ in range(vertices)]
        self.nodes = {}

    def add_vertex(self, value, index):
        self.nodes[value] = index

    def add_edge(self, u, v):

        i = self.nodes[u]
        j = self.nodes[v]

        self.matrix[i][j] = 1
        self.matrix[j][i] = 1

    def display(self):

        for row in self.matrix:
            print(row)


g = Graph(4)

g.add_vertex(10,0)
g.add_vertex(20,1)
g.add_vertex(30,2)
g.add_vertex(40,3)

g.add_edge(10,20)
g.add_edge(10,30)
g.add_edge(20,40)
g.add_edge(30,40)

g.display()
```

```python
# bfs

from collections import deque


class Graph:
    
    def __init__(self):
        self.graph = {}
        
    def add_edge(self, u, v):
        
        if u not in self.graph:
            self.graph[u] = []
            
        if v not in self.graph:
            self.graph[v] = []
            
        self.graph[u].append(v)
        
    def bfs(self, start):
        
        visited = set()
        queue = deque()
        queue.append(start)
        visited.add(start)
        
        while queue:
            node = queue.popleft()
            print(node, end = " ")
            
            for neighbor in self.graph.get(node, []):
                if neighbor not in visited:
                    visited.add(neighbor)
                    queue.append(neighbor)
                    
g = Graph()
g.add_edge(0,1)
g.add_edge(0,2)
g.add_edge(1,2)
g.add_edge(2,0)
g.add_edge(2,3)
g.add_edge(3,3)

print("BFS starting from node 0:")

g.bfs(0)
```

```python
class Graph:
    
    def __init__(self):
        self.graph = {}

    def add_edge(self, u, v):

        if u not in self.graph:
            self.graph[u] = []

        self.graph[u].append(v)

    def dfs(self, node, visited):

        visited.add(node)
        print(node, end=" ")

        for neighbor in self.graph.get(node, []):

            if neighbor not in visited:
                self.dfs(neighbor, visited)



g = Graph()

g.add_edge(0,1)
g.add_edge(0,2)
g.add_edge(1,3)
g.add_edge(1,4)

print("DFS traversal starting from node 0:")

visited = set()
g.dfs(0, visited)
```
